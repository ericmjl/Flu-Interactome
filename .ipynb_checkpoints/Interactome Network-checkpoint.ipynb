In this notebook, I aim to create the interactome network from the Harvard CCSB Human Binary Interactome dataset.

http://interactome.dfci.harvard.edu/H_sapiens/index.php?page=download

The dataset involves a list of proteins that are known to interact with one another, based on their high-throughput screening experiments. It is simply a binary yes/no list of interactions; direction, strength, and other important data are not available.

```python
import pandas as pd
import networkx as nx

dataset = pd.read_csv('HI-II-14.tsv', delimiter = '\t')
dataset
```

```python
# Construct the network.
G = nx.Graph()
for row, data in dataset.iterrows():
    G.add_edge(data['Symbol A'], data['Symbol B'])
    
    G.node[data['Symbol A']]['gene_id'] = data['Entrez Gene IDA']
    G.node[data['Symbol B']]['gene_id'] = data['Entrez Gene IDB']
```

```python
G.nodes(data=True)[0]
```

```python
[n for n, d in G.nodes(data=True) if 'NFKB' in n]
```

```python
G.neighbors('ACTB')
```

```python
sorted(nx.degree_centrality(G).items(), key=lambda x: x[1], reverse=True)
```
